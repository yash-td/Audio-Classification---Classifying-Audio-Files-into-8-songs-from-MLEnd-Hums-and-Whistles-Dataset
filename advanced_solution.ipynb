
In the next cell, we define a new function that gets the pitch of an audio signal (do not worry if you do not know what it is, but feel free to read about it!).

```python
def getPitch(x,fs,winLen=0.02):
  #winLen = 0.02 
  p = winLen*fs
  frame_length = int(2**int(p-1).bit_length())
  hop_length = frame_length//2
  f0, voiced_flag, voiced_probs = librosa.pyin(y=x, fmin=80, fmax=450, sr=fs,
                                                 frame_length=frame_length,hop_length=hop_length)
  return f0,voiced_flag
```

Here we have created some functions which convert the audio signal into a midi value with musical scales. Later in our pipeline we transform this musical scales to corresponding log values of its frequency in order to pass them as a feature to our model. Some of the functions used are directly imported from librosa (ones to convert note to midi or note to hz)

```python
# Extracting the MIDI values of Files
!pip install midiutil
import numpy as np
import librosa
import midiutil
import sys


def transition_matrix(note_min, note_max, p_stay_note, p_stay_silence):
    midi_min = librosa.note_to_midi(note_min)
    midi_max = librosa.note_to_midi(note_max)
    n_notes = midi_max - midi_min + 1
    p_ = (1-p_stay_silence)/n_notes
    p__ = (1-p_stay_note)/(n_notes+1)
    
    # Transition matrix:
    # State 0 = silence
    # States 1, 3, 5... = onsets
    # States 2, 4, 6... = sustains
    T = np.zeros((2*n_notes+1, 2*n_notes+1))

    # State 0: silence
    T[0,0] = p_stay_silence
    for i in range(n_notes):
        T[0, (i*2)+1] = p_
    
    # States 1, 3, 5... = onsets
    for i in range(n_notes):
        T[(i*2)+1, (i*2)+2] = 1

    # States 2, 4, 6... = sustains
    for i in range(n_notes):
        T[(i*2)+2, 0] = p__
        T[(i*2)+2, (i*2)+2] = p_stay_note
        for j in range(n_notes):        
            T[(i*2)+2, (j*2)+1] = p__
    
    return T


def probabilities(y, note_min, note_max, sr, frame_length, window_length, hop_length, pitch_acc, voiced_acc, onset_acc, spread):

    fmin = librosa.note_to_hz(note_min)
    fmax = librosa.note_to_hz(note_max)
    midi_min = librosa.note_to_midi(note_min)
    midi_max = librosa.note_to_midi(note_max)
    n_notes = midi_max - midi_min + 1
    
    # F0 and voicing
    f0, voiced_flag, voiced_prob = librosa.pyin(y, fmin*0.9, fmax*1.1, sr, frame_length, window_length, hop_length)
    tuning = librosa.pitch_tuning(f0)
    f0_ = np.round(librosa.hz_to_midi(f0-tuning)).astype(int)
    onsets = librosa.onset.onset_detect(y, sr=sr, hop_length=hop_length, backtrack=True)


    P = np.ones( (n_notes*2 + 1, len(f0)) )

    for t in range(len(f0)):
        if voiced_flag[t]==False:
            P[0,t] = voiced_acc
        else:
            P[0,t] = 1-voiced_acc

        for j in range(n_notes):
            if t in onsets:
                P[(j*2)+1, t] = onset_acc
            else:
                P[(j*2)+1, t] = 1-onset_acc

            if j+midi_min == f0_[t]:
                P[(j*2)+2, t] = pitch_acc

            elif np.abs(j+midi_min-f0_[t])==1:
                P[(j*2)+2, t] = pitch_acc * spread

            else:
                P[(j*2)+2, t] = 1-pitch_acc

    return P,f0, voiced_flag

def states_to_pianoroll(states, note_min, note_max, hop_time):

    midi_min = librosa.note_to_midi(note_min)
    midi_max = librosa.note_to_midi(note_max)
    
    states_ = np.hstack( (states, np.zeros(1)))
    
    # possible types of states
    silence = 0
    onset = 1
    sustain = 2

    my_state = silence
    output = []
    
    last_onset = 0
    last_offset = 0
    last_midi = 0
    for i in range(len(states_)):
        if my_state == silence:
            if int(states_[i]%2) != 0:
                # Found an onset!
                last_onset = i * hop_time
                last_midi = ((states_[i]-1)/2)+midi_min
                last_note = librosa.midi_to_note(last_midi)
                my_state = onset


        elif my_state == onset:
            if int(states_[i]%2) == 0:
                my_state = sustain

        elif my_state == sustain:
            if int(states_[i]%2) != 0:
                # Found an onset.                
                # Finish last note
                last_offset = i*hop_time
                my_note = [last_note]
                output.append(my_note)
                
                # Start new note
                last_onset = i * hop_time
                last_midi = ((states_[i]-1)/2)+midi_min
                last_note = librosa.midi_to_note(last_midi)
                my_state = onset
            
            elif states_[i]==0:
                # Found silence. Finish last note.
                last_offset = i*hop_time
                my_note = [last_note]
                output.append(my_note)
                my_state = silence

    return output
```

###  Extracting the features from our data

Creating a function to extract the features that we finalised after testing different various permutations and comibations. Here we are adding an extra feature (log values of musical note frequencies).Further this function will append labels for 8 different types of songs that we are going to classify. Hence creating 8 classes for the classification probelm. The label matrix will be an array with 8 unique values indicating a one of the 8 songs.

```python
def getXy(files,labels_file, scale_audio=False, onlySingleDigit=False):


  #setting values to extract the midi values
  note_min='A2'
  note_max='E6'
  voiced_acc = 0.9
  onset_acc = 0.8
  frame_length=2048
  window_length=1024
  hop_length=256
  pitch_acc = 0.99
  spread = 0.6
  
    
  X,y =[],[]
  Xi = []
  for file in tqdm(files):
    fileID = file.split('/')[-1]
    file_name = file.split('/')[-1]\

    if labels_file.loc[fileID]['song']=='Potter' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Potter'
    if labels_file.loc[fileID]['song']=='StarWars' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'StarWars'
    if labels_file.loc[fileID]['song']=='Panther' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Panther'
    if labels_file.loc[fileID]['song']=='Showman' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Showman'  
    if labels_file.loc[fileID]['song']=='Mamma' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Mamma'
    if labels_file.loc[fileID]['song']=='Hakuna' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Hakuna'
    if labels_file.loc[fileID]['song']=='Rain' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Rain'
    if labels_file.loc[fileID]['song']=='Frozen' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Frozen'

    fs = None # if None, fs would be 22050
    x, fs = librosa.load(file, res_type='kaiser_fast')
 
    T = transition_matrix(note_min, note_max, 0.9, 0.2)
    P,f0,voiced_flag  = probabilities(x, note_min, note_max, fs, frame_length, window_length, hop_length, pitch_acc, voiced_acc, onset_acc, spread)
    p_init = np.zeros(T.shape[0])
    p_init[0] = 1
    
    states = librosa.sequence.viterbi(P, T, p_init=p_init)
    pianoroll=states_to_pianoroll(states, note_min, note_max, hop_length/fs)
    pianoroll = np.array(pianoroll)
    power = np.sum(x**2)/len(x)
    pitch_mean = np.nanmean(f0) if np.mean(np.isnan(f0))<1 else 0
    pitch_std  = np.nanstd(f0) if np.mean(np.isnan(f0))<1 else 0
    voiced_fr = np.mean(voiced_flag)
    mfccs = np.mean(librosa.feature.mfcc(y=x, sr=fs, n_mfcc=13).T,axis=0)
    cent  = np.mean(librosa.feature.spectral_centroid(y=x, sr=fs).T,axis=0)
    mfccs = np.append(mfccs,cent)
    xi = [power,pitch_mean,pitch_std,voiced_fr]
    chromagram = np.mean(librosa.feature.chroma_stft(x, sr=fs, hop_length=hop_length).T)
    mfccs = np.append(mfccs,chromagram)
    mel = np.mean(librosa.feature.melspectrogram(x,sr=fs).T,axis=0)
    mfccs = np.append(mfccs,mel)
    xi = np.append(mfccs,xi)
    X.append(pianoroll)
    lenmin = min(map(len, X)) #calculating thye length of midi list with minimmum number of valiues in order to make the dimension of our training data uniform
    for i in range(len(X)):  #setting the maximum midi length of all the files equal to that of the midi list with minimum values
      if len(X) != 0:
        X[i] = X[i][:lenmin]

    Xi.append(xi)
    y.append(yi)
  return np.array(X),np.array(y),np.array(Xi)
```

```python
# Training our model without the log values of musical scale frequencies.

def getXy_new(files,labels_file, scale_audio=False, onlySingleDigit=False):
  note_min='A2'
  note_max='E6'
  voiced_acc = 0.9
  onset_acc = 0.8
  frame_length=2048
  window_length=1024
  hop_length=256
  pitch_acc = 0.99
  spread = 0.6
  
    
  X,y =[],[]
  Xi = []
  for file in tqdm(files):
    fileID = file.split('/')[-1]
    file_name = file.split('/')[-1]\

    if labels_file.loc[fileID]['song']=='Potter' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Potter'
    if labels_file.loc[fileID]['song']=='StarWars' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'StarWars'
    if labels_file.loc[fileID]['song']=='Panther' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Panther'
    if labels_file.loc[fileID]['song']=='Showman' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Showman'  
    if labels_file.loc[fileID]['song']=='Mamma' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Mamma'
    if labels_file.loc[fileID]['song']=='Hakuna' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Hakuna'
    if labels_file.loc[fileID]['song']=='Rain' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Rain'
    if labels_file.loc[fileID]['song']=='Frozen' and (labels_file.loc[fileID]['interpretation']=='hum' or labels_file.loc[fileID]['interpretation']=='whistle'):
      yi = 'Frozen'

    fs = None # if None, fs would be 22050
    x, fs = librosa.load(file, res_type='kaiser_fast')
    if scale_audio: x = x/np.max(np.abs(x))
    f0, voiced_flag = getPitch(x,fs,winLen=0.02)
    power = np.sum(x**2)/len(x)
    pitch_mean = np.nanmean(f0) if np.mean(np.isnan(f0))<1 else 0
    pitch_std  = np.nanstd(f0) if np.mean(np.isnan(f0))<1 else 0
    voiced_fr = np.mean(voiced_flag)
    mfccs = np.mean(librosa.feature.mfcc(y=x, sr=fs, n_mfcc=13).T,axis=0)
    cent  = np.mean(librosa.feature.spectral_centroid(y=x, sr=fs).T,axis=0)
    mfccs = np.append(mfccs,cent)
    xi = [power,pitch_mean,pitch_std,voiced_fr]
    mfccs = np.append(mfccs,xi)
    chromagram = np.mean(librosa.feature.chroma_stft(x, sr=fs, hop_length=hop_length).T)
    mfccs = np.append(mfccs,chromagram)
    mel = np.mean(librosa.feature.melspectrogram(x,sr=fs).T,axis=0)
    mfccs = np.append(mfccs,mel)
    X.append(mfccs)
    y.append(yi)
  return np.array(X),np.array(y)
```
